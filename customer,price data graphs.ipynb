```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customer_data = pd.read_csv(r"C:\Users\Fathimath Sareena\Downloads\client_data (1).csv")
price_data = pd.read_csv(r"C:\Users\Fathimath Sareena\Downloads\price_data (1).csv")
```

```python
print("Customer Data Types:\n", customer_data.dtypes)
print("\nPrice Data Types:\n", price_data.dtypes)
```

```python
numerical_columns_customer = customer_data.select_dtypes(include=['float64', 'int64']).columns
```

```python
numerical_columns_price = price_data.select_dtypes(include=['float64', 'int64']).columns
```

```python
for column in numerical_columns_customer:
    plt.figure(figsize=(8, 4))
    sns.histplot(customer_data[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
for column in numerical_columns_price:
    plt.figure(figsize=(8, 4))
    sns.histplot(price_data[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
# Alternative visualizations
for column in numerical_columns_customer:
    plt.figure(figsize=(12, 6))
    
    # Boxplot
    plt.subplot(1, 2, 1)
    sns.boxplot(data=customer_data[column])
    plt.title(f'Box Plot of {column}')
    
    # Density plot
    plt.subplot(1, 2, 2)
    sns.kdeplot(customer_data[column], shade=True)
    plt.title(f'Density Plot of {column}')
    
    plt.tight_layout()
    plt.show()
```
